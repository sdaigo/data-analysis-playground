# 度数分布表

```python
data = {
    "A": [1, 10, 1, 10, 1, 10, 1,10],
    "B": [5, 5, 5, 5, 6, 6, 6, 6],
    "C": [1, 2, 3, 4, 7, 8, 9, 10],
}

df = pd.DataFrame(data)
df
```

```python
df.mean()
```

```python
df.median()
```

```python
df.mode()
```

```python
def freq_distribution(col):
    bins = [1,3,5,7,9,11]
    cut = pd.cut(df[col], bins=bins, right=False)
    return cut.value_counts(sort=False)

freq_distribution("A")
```

```python
freq_distribution("B")
```

```python
freq_distribution("C")
```
