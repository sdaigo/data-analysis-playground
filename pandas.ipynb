# 欠損値の処理

```python
data = {
    "国語": [90, 50, None, 40],
    "数学": [80, None, None, 50]
}

idx = ["Kid A", "Kid B", "Kid C", "Kid D"]
dfA = pd.DataFrame(data, index=idx)

dfA
```

```python
# 欠損値の個数
dfA.isnull().sum()
```

```python
# 除去
dfA.dropna()
```

```python
# 国語に欠損値がある行だけ除去する
dfA.dropna(subset=["国語"])
```

```python
# 欠損値を平均値で埋める
dfB = dfA.fillna(dfA.mean())
dfB
```

```python
# 一つ前の値で埋める
dfB = dfA.fillna(method="ffill")
dfB
```

# 重複したデータを除去する

```python
data = [
    [10, 30, 40],
    [20, 30, 40], # dup
    [20, 30, 40], # dup
    [30, 30, 50],
    [20, 30, 40], # dup
]

dfA = pd.DataFrame(data)
dfA
```

```python
dfA.duplicated().value_counts()
```

```python
dfA.drop_duplicates()
```

# 文字列を数値に直す

```python
data = {
    "A": ["100", "300"],
    "B": ["500", "1,500"]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.dtypes
```

```python
df["A"] = df["A"].astype(int)
```

```python
df
```

```python
df["B"] = df["B"].str.replace(",", "").astype(int)
```
